##### 7. Cek outlier data

```python
# Menghitung total seluruh baris data sebagai pembagi persentase
total_data = df.shape[0]

print("=== DETAIL BATAS & PERSENTASE OUTLIER ===\n")

# ================================================
# 1. ANALISIS OUTLIER KOLOM: Acquisition_Cost (AC)
# ================================================
Q1_AC = df['Acquisition_Cost'].quantile(0.25)
Q3_AC = df['Acquisition_Cost'].quantile(0.75)
IQR_AC = Q3_AC - Q1_AC
lower_AC = Q1_AC - (1.5 * IQR_AC)
upper_AC = Q3_AC + (1.5 * IQR_AC)

# Menghitung jumlah baris yang di luar batas
outlier_AC = df[(df['Acquisition_Cost'] < lower_AC) | (df['Acquisition_Cost'] > upper_AC)].shape[0]
# Rumus persentase: (jumlah outlier / total data) * 100
pct_AC = (outlier_AC / total_data) * 100

print(f"--- Kolom Acquisition_Cost ---")
print(f"Batas Bawah: {lower_AC:.2f} | Batas Atas: {upper_AC:.2f}")
print(f"Jumlah Outlier: {outlier_AC} baris")
print(f"Persentase Outlier: {pct_AC:.2f}%\n")


# ==================================
# 2. ANALISIS OUTLIER KOLOM: Clicks
# ==================================
Q1_Clicks = df['Clicks'].quantile(0.25)
Q3_Clicks = df['Clicks'].quantile(0.75)
IQR_Clicks = Q3_Clicks - Q1_Clicks
lower_Clicks = Q1_Clicks - (1.5 * IQR_Clicks)
upper_Clicks = Q3_Clicks + (1.5 * IQR_Clicks)

outlier_Clicks = df[(df['Clicks'] < lower_Clicks) | (df['Clicks'] > upper_Clicks)].shape[0]
pct_Clicks = (outlier_Clicks / total_data) * 100

print(f"--- Kolom Clicks ---")
print(f"Batas Bawah: {lower_Clicks:.2f} | Batas Atas: {upper_Clicks:.2f}")
print(f"Jumlah Outlier: {outlier_Clicks} baris")
print(f"Persentase Outlier: {pct_Clicks:.2f}%\n")


# ==========================================
# 3. ANALISIS OUTLIER KOLOM: Impressions
# ==========================================
Q1_Impressions = df['Impressions'].quantile(0.25)
Q3_Impressions = df['Impressions'].quantile(0.75)
IQR_Impressions = Q3_Impressions - Q1_Impressions
lower_Impressions = Q1_Impressions - (1.5 * IQR_Impressions)
upper_Impressions = Q3_Impressions + (1.5 * IQR_Impressions)

outlier_Impressions = df[(df['Impressions'] < lower_Impressions) | (df['Impressions'] > upper_Impressions)].shape[0]
pct_Impressions = (outlier_Impressions / total_data) * 100

print(f"--- Kolom Impressions ---")
print(f"Batas Bawah: {lower_Impressions:.2f} | Batas Atas: {upper_Impressions:.2f}")
print(f"Jumlah Outlier: {outlier_Impressions} baris")
print(f"Persentase Outlier: {pct_Impressions:.2f}%\n")


# ================================================
# 4. ANALISIS OUTLIER KOLOM: Engagement_Score (ES)
# ================================================
Q1_ES = df['Impressions'].quantile(0.25)
Q3_ES = df['Impressions'].quantile(0.75)
IQR_ES = Q3_ES - Q1_ES
lower_ES = Q1_ES - (1.5 * IQR_ES)
upper_ES = Q3_ES + (1.5 * IQR_ES)

outlier_ES = df[(df['Impressions'] < lower_ES) | (df['Impressions'] > upper_ES)].shape[0]
pct_ES = (outlier_ES / total_data) * 100

print(f"--- Kolom Engagement_Score ---")
print(f"Batas Bawah: {lower_ES:.2f} | Batas Atas: {upper_ES:.2f}")
print(f"Jumlah Outlier: {outlier_ES} baris")
print(f"Persentase Outlier: {pct_ES:.2f}%\n")
```

### 4. Data Manipulation

##### 1. Membuat kolom baru melalui Feature Engineering

##### a. Membuat kolom Duration_Days

```python
# Membuat kolom Duaration_Days hanya berisi angka 
# Tujuannya untuk digunakan dalam visualisasi dan typenya integer

df['Duration_Days'] = df['Duration'].str.replace('days', '', case=False).astype(int)
```

##### b. Membuat kolom Cost_per_Click (CPC)

```python
# Tujuan : untuk mengetahui biaya yang dikeluarkan perusahaan untuk setiap klik iklan

df['CPC'] = (df['Acquisition_Cost'] / df['Clicks']).round(2)
```

##### c. Membuat kolom Click_Through_Rate (CTR)

```python
# Tujuan : untuk mengukur efktivitas materi iklan (seberapa banyak orang yang mengklik setelah iklan tersebut muncul)
# Satuannya adalah persen maka dikalikan 100

df['CTR'] = ((df['Clicks'] / df['Impressions'])*100).round(2)
```

##### d. Membuat kolom nama bulan

```python
# Tujuan : untuk melihat bulan dalam dataset dan nantinya dianalisis untuk melihat performa data setiap bulannya

df['Month'] = df['Date'].dt.strftime('%B')
```

##### e. Membuat kolom kategori performa ROI

```python
# Tujuan : Mengelolompokkan kampanye mana yang bernilai boncos, standar, atau untung besar
# Dibuat kategori ROI menjadi 3 yaitu nilai ROI<3 (Low), 3-6 (Medium), >6 (High)

def nilai_roi(x):
    if x < 3.0:
        return 'Low'
    elif x <= 6.0:
        return 'Medium'
    else:
        return 'High'

df['ROI_Performance'] = df['ROI'].apply(nilai_roi)
```

```python
print(df['ROI_Performance'].value_counts())
```

##### f. Membuat kolom kategori durasi waktu iklan

```python
# Tujuan : untuk mengelompokkan mana iklan dengan waktu tayang yang jangka pendek, menengah, atau panjang
# Bisa digunakan untuk menganalisis dengan biaya-biaya yang dikeluarkan perusahaan untuk sebuah promotion/iklan

def kategori_durasi(x):
    if x == 15:
        return 'Short-term'
    elif x == 30 or x == 45:
        return 'Medium-term'
    else:
        return 'Long-term'

df['Duration_Category'] = df['Duration_Days'].apply(kategori_durasi)
```
